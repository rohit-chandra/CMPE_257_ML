```python
import pandas as pd 
import numpy as np
from numpy import loadtxt
from sklearn.model_selection import train_test_split
from keras.models import Sequential
from keras.layers import Dense, BatchNormalization, Dropout
from sklearn.metrics import f1_score

from google.colab import files
files.upload()

data = pd.read_csv('hcc-data-complete-balanced.csv')
print(data[0:3])
```

Some columns have comma separted values 

```python
data.iloc[5,25]
```

The dataset has no null values and hence there won't be any replacement 

```python
data.isnull().sum()  # no null values are present 
```

Dropping duplicates from the data to clean it. But it was found that data doesn't have duplicates.

```python
data.drop_duplicates()  # no duplicates found 
```

Splitting the dataset into attributes(X) and target(Y) features

```python
X = data.iloc[:,:-1]
Y = data.iloc[:,-1]

X = pd.DataFrame(X)
print(Y)
print(X.shape)
```

Data Cleaning step ==> Replacing the comma with decimal for columns having continuous data

```python
X = X.replace(',','.', regex=True)
print(X.iloc[5,25])
X = X.astype('float32')
X.shape
```

Performing Feature Engineering to add new features such as Anaemic and Unconjugated(Indirect) Bilirubin from existing features. This involves deep domain knowledge, due to restricted domain knowledge I just added two simple columns.

```python
X['Indir_Bil'] = X['Total_Bil']-X['Dir_Bil']
X['Anaemic'] = [1 if i<45 else 0 for i in X['Ferritin'] ]

X.shape
```

Splitting the dataset into training and test set



```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size=0.3)
X_train.shape
```

Building a Neural Network model

```python
model = Sequential()
model.add(Dense(200,input_dim=51, activation='relu', use_bias=True))   # dense implements activation( input.kernel + bias)
model.add(Dense(100, activation='relu', use_bias=True))
model.add(Dense(150, activation='relu', use_bias=True))
model.add(Dense(100, activation='relu', use_bias=True))
model.add(Dense(100, activation='relu', use_bias=True))
model.add(Dense(1,activation='sigmoid', use_bias=True))

model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])

model.fit(X_train,Y_train, epochs=100, batch_size=20, verbose=0)

loss,accuracy = model.evaluate(X_train,Y_train)
print("Loss = ", loss, "Accuracy = ",accuracy)
```

Relu and Sigmoid activation function were used.
1. ReLU does not activate all the neurons at the same time hence it is computationally efficient.
2. Because the neurons will be deactivated if the output of linear transformation is less than 0
3. ReLU accelerates the convergence of gradient descent towards global minimum of loss function
4. Since output should be either 0 or 1, Sigmoid activation function is used in the output layer.
5. Sigmoid provides a smooth curve and prevent jumping of output values 

```python
y_pred = model.predict(X_test, verbose=0)
y_pred = (y_pred> 0.5).astype("int32")
print("F1 Score = {:.2f}".format(f1_score(Y_test,y_pred)))
```

HyperParameter tuning using GridSearchCV and KerasClassifier wrapper class

```python
import numpy
from sklearn.model_selection import GridSearchCV
from keras.wrappers.scikit_learn import KerasClassifier

def build_clf():
  model = Sequential()
  model.add(Dense(200,input_dim=51, activation='relu', use_bias=True))   # dense implements activation( input.kernel + bias)
  model.add(Dense(100, activation='relu', use_bias=True))
  model.add(Dense(150, activation='relu', use_bias=True))
  model.add(Dense(100, activation='relu', use_bias=True))
  model.add(Dense(100, activation='relu', use_bias=True))
  model.add(Dense(1,activation='sigmoid', use_bias=True))
  model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])

  return model

model=KerasClassifier(build_fn=build_clf)

params={'batch_size':[100, 20, 50, 25, 32], 'epochs':[200, 100, 300, 400] }

gs = GridSearchCV(estimator=model, param_grid=params, cv=10, scoring='accuracy') 
gs = gs.fit(X_train, Y_train)
```

```python
best_params=gs.best_params_
print("After hyperparameter tuning we found these paramaters to be most precise ",best_params)

best_accuracy=gs.best_score_
print(best_accuracy)
```

Adding Batch Normalization

```python
model = Sequential()
model.add(Dense(200,input_dim=51, activation='relu', use_bias=True))
BatchNormalization()
model.add(Dense(100, activation='relu', use_bias=True))
BatchNormalization()
model.add(Dense(150, activation='relu', use_bias=True))
BatchNormalization()
model.add(Dense(100, activation='relu', use_bias=True))
BatchNormalization()
model.add(Dense(100, activation='relu', use_bias=True))
BatchNormalization()
model.add(Dense(1,activation='sigmoid'))

model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])

model.fit(X_train,Y_train, epochs=100, batch_size=20)

print(model.evaluate(X_train,Y_train))
```

```python
_, accuracy = model.evaluate(X_test, Y_test)
print(accuracy)
```

After adding Dropout after every layer the accuracy dropped.

```python
y_pred = model.predict(X_test, verbose=0)
y_pred = (y_pred> 0.5).astype("int32")
print("F1 Score = {:.2f}".format(f1_score(Y_test,y_pred)))
```

```python
model = Sequential()
model.add(Dense(200,input_dim=51, activation='relu', use_bias=True))
model.add(Dropout(0.3))
model.add(Dense(100, activation='relu', use_bias=True))
model.add(Dropout(0.3))
model.add(Dense(150, activation='relu', use_bias=True))
model.add(Dropout(0.3))
model.add(Dense(100, activation='relu', use_bias=True))
model.add(Dropout(0.3))
model.add(Dense(100, activation='relu', use_bias=True))
model.add(Dropout(0.3))
model.add(Dense(1,activation='sigmoid'))

model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])

model.fit(X_train,Y_train, epochs=100, batch_size=20)

print(model.evaluate(X_train,Y_train))
```

```python
_, accuracy = model.evaluate(X_test, Y_test)
print(accuracy)
```

```python
y_pred = model.predict(X_test, verbose=0)
y_pred = (y_pred> 0.5).astype("int32")
print("F1 Score = {:.2f}".format(f1_score(Y_test,y_pred)))
```
