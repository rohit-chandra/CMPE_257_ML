Reading movies.dat and ratings.dat file and converting them into a DataFrame

```python
import pandas as pd 
import numpy as np

from google.colab import files
files.upload()

from google.colab import files
files.upload()

movies = pd.DataFrame(pd.read_table('movies.dat', sep='::', header=None, engine='python', encoding='ISO-8859-1', names=['movieId', 'movie', 'genre']))
print(movies.head())

ratings = pd.DataFrame(pd.read_table('ratings.dat', sep='::', header=None, engine='python', encoding='ISO-8859-1', names=['userId', 'movieId', 'rating', 'time']))
print(ratings.head())
```

Creating "ratings" matrix where **movie ID are rows** and **user ID as columns** and the values are the rating given by that user to that particular movie

```python
ratings = ratings.pivot(index='movieId',columns='userId',values='rating')
ratings.head()
```

Replacing NaN values with 0 to increase readability of the data 

```python
ratings.fillna(0,inplace=True)
ratings.head()
```

Normalizing the ratings matrix to avoid the effect of 

```python
normalised_mat = ratings - np.asarray([(np.mean(ratings,1))]).T
print(normalised_mat.head())
print(normalised_mat.shape)
```

Using np.linalg.svd to decompose the normalized matrix into component matrices U, S and V

```python
A = normalised_mat.T / np.sqrt(ratings.shape[0] - 1)
U, S, V = np.linalg.svd(A) #SVD is used for collaborative filtering 
print(U.shape)
print(S.shape)
print(V.shape)
```

cos_sim function finds the cosine similarity between two movies i.e two rows and return the top similar movies based on cosine similarity score. print_movies function takes the indexes returned by the cos_sim function and print the movie's name from the original movie data's matrix i.e "movies"

```python
def cos_sim(data, movie_id, top_n=10):
    index = movie_id - 1 # Movie id starts from 1
    movie_row = data[index, :]
    magnitude = np.sqrt(np.einsum('ij, ij -> i', data, data))
    similarity = np.dot(movie_row, data.T) / (magnitude[index] * magnitude)
    sort_indexes = np.argsort(-similarity)
    return sort_indexes[:top_n]

def print_movies(movies, movie_id, top_indexes):
    print('Recommendations for {0}: \n'.format(
    movies[movies.movieId == movie_id].movie.values[0]))
    for id in top_indexes + 1:
        print(movies[movies.movieId == id].movie.values[0])
```

```python
k = 50
movie_id = 7 # Grab an id from movies.dat
top = 10

sliced = V.T[:, :k] # selecting only 50 components of V transpose 
indexes = cos_sim(sliced, movie_id, top)
print_movies(movies, movie_id, indexes)
```

Finding Eigen values and eigen vectors of the covariance matrix of the normalized ratings matrix and using them to find Similar movies

```python
cov_mat = np.cov(normalised_mat)
evals, evecs = np.linalg.eig(cov_mat)
```

```python
sliced = evecs[:, :k] # slicing to use only 50 eigen vectors 
top_indexes = cos_sim(sliced, movie_id, top)
print_movies(movies, movie_id, top_indexes)
```
