```python
#finding youngest person

youngest_person=train_df["age"].min()
youngest_person

print("The youngest person is ",format(youngest_person),"years old.")
```

```python
#finding average age group

average_person=train_df["age"].mean()
average_person

print("The average age group is ",format(average_person),"years old.")
```

```python
#finding median age

med_person=train_df["age"].median()
med_person

print("The median age is ",format(med_person),"years old.")
```

# E

```python
#finding relationship between Cardio And Ages
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.kdeplot(train_df.age[train_df.cardio==1],color="darkturquoise",shade=True)

sns.kdeplot(train_df.age[train_df.cardio==0],color="lightcoral",shade=True)

plt.legend(['age','cardio disease'])
plt.title('age vs cardio disease')
plt.xlim(30,80)
plt.show()
```

We observe that most of the cardiovascular disease are seen in the people of age group 40 to 60.
